```python
#wap to display all positions of the substring in the given main string

s1=input('enter the main string: ')
s2=input('enter the substring: ')
n=len(s1)
pos=0
count=0
while True:
    i=s1.find(s2,pos,n)
    if i==-1:
        break
    else:
        print("Occured at index ",i)
        count+=1
        pos=i+len(s2)
print("No of Occurances", count)
```

```python
s='abcdef'
even=[]
odd=[]
i=0

while i<len(s):
    if i%2==0:
        even.append(s[i])
    else:
        odd.append(s[i])
    i=i+1

print('characters present at even index', even)
print('characters present at odd index', odd)
```

```python
#wap to merge characters of two strings into a single string by taaking characters alternatively 

s1='ritik'
s2='mishra'
op=''
i=0
j=0

while i<len(s1) or j<len(s2):
    op = op + s1[i] + s2[j]
    i=i+1
    j=j+1
print(op)
```

```python
s1 = 'ritik'
s2 = 'mishra'
op = ''
i = 0
j = 0

while i < len(s1) or j < len(s2):
    if i < len(s1):       # Check before accessing s1[i]
        op = op + s1[i]
        i += 1
    if j < len(s2):       # Check before accessing s2[j]
        op = op + s2[j]
        j += 1

print(op)
```
