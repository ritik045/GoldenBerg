```python
n=int(input('enter the no of rows'))
st=1 #stars
sp=n//2  #spaces
i=0
while i<n:
    for j in range(sp):
        print(' ', end=' ')
    for j in range(st):
        print('*', end=' ')
    print()
    if i<n//2:
        st+=2
        sp-=1
    else:
        st-=2
        sp+=1
    i+=1
    print()
```
