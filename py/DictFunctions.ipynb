```python
x={10,20,30,40}
y={10,20,50,60}
print(x.union(y))
print(x | y)
```

```python
x={10,20,30,40}
y={10,20,50,60}
print(x.intersection(y))
print(x&y)
```

```python
x={10,20,30,40}
y={10,20,50,60}
print(x.difference(y))
print(x-y)
```

```python
x={10,20,30,40}
y={10,20,50,60}
print(x.symmetric_difference(y))
print(x ^ y)
```

```python
# comprehension

s={x*x for x in range(1,6) if x%2==0}
print(s)
print(type(s))
```
